```python
import matplotlib.pyplot as plt
```

```python
x =  data["year"]
y =  data["unemployed"]
y4 = data["unemployed_percent"]
y3 = data["% real"]
y2 = data["Real"]
y1 = data["Nominal"]
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Unemployment')
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Nominal GDP')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Unemployment')
axs[0, 1].plot(x, y2, 'tab:red')
axs[0, 1].set_title('Real GDP')
axs[1, 0].plot(x, y4, 'tab:green')
axs[1, 0].set_title('Unemployment rate')
axs[1, 0].plot(x, y3, 'tab:blue')
axs[1, 0].set_title('Real GDP rate')
fig.tight_layout(pad=3.0)
```
